```python
import pandas as pd

df = pd.read_excel('datasets/FoodKeeper-Data.xls', sheet_name = 'Product')

def isNaN(string): return string != string

def keywordFinder(key):
    # print(df['Keywords'].iloc[4])
    for index  in range(len(df['Keywords'])-1):
        keywordBank = df['Keywords'].iloc[index]
        if isNaN(keywordBank) == False:
            temp = keywordBank.split(",")
            for keys in temp:
                if keys.lower() == key.lower():
                    print(keywordBank.lower(), index)
                    print("Name: ", df['Name'].iloc[index])


def foodStorage(dfName, foodName, information):
    '''
    1. create enum list of pantry, ref, freeze
        - give user all info...
    2. use switch or match to make the code more readable
    3. create a list of possible responses back to users,
        - Ex. "can be stored in the pantry for..."
    4. If we find the food, check whether it has any of the column info.
        - have a temp list of all possible dictonary columns, checking if isNaN
    5. return empty string '', utilizng with Brandon's code.
    
    '''
    
    storageInfo = {
        "Pantry": [
            "Pantry_Min",
            "Pantry_Max",
            "Pantry_Metric",
            "Pantry_tips",
            "DOP_Pantry_Min",
            "DOP_Pantry_Max",
            "DOP_Pantry_Metric",
            "DOP_Pantry_tips",
            "Pantry_After_Opening_Min",
            "Pantry_After_Opening_Max",
            "Pantry_After_Opening_Metric"
        ],
        "Refrigerate": [
            'Refrigerate_Min',
            'Refrigerate_Max',
            'Refrigerate_Metric',
            'Refrigerate_tips',
            'DOP_Refrigerate_Min',
            'DOP_Refrigerate_Max',
            'DOP_Refrigerate_Metric',
            'DOP_Refrigerate_tips',
            'Refrigerate_After_Opening_Min',
            'Refrigerate_After_Opening_Max',
            'Refrigerate_After_Opening_Metric',
            'Refrigerate_After_Thawing_Min',
            'Refrigerate_After_Thawing_Max',
            'Refrigerate_After_Thawing_Metric',
            'Refrigerate_Min',
        ],
        "Freeze" : [
            "Freeze_Min",
            "Freeze_Max",
            "Freeze_Metric",
            "Freeze_Tips",
            "DOP_Freeze_Min",
            "DOP_Freeze_Max",
            "DOP_Freeze_Metric",
            "DOP_Freeze_Tips"
        ]
    }

    index = 431
    tempCol = []
    tempString = ''
    foodName = df["Name"].iloc[index]
    info = information.lower().capitalize()

    if storageInfo.get(info):
        # print(storageInfo[info])
        # print(df["Name"].iloc[index])
        for column in storageInfo[info]:
            if isNaN(df[column].iloc[index]) == False:
                tempCol.append(column)

    
    for i in ['Min', 'DOP', 'After', 'tips', 'Tips']:
        for col in tempCol:
            if i in col:
                tempString = ''
                if i == "Min" and i != "DOP":
                    min = df[info + '_Min'].iloc[index]
                    max = df[info + '_Max'].iloc[index]
                    if min == max: tempString += f'{foodName} can be stored in the {info} for {min} {df[info + "_Metric"].iloc[index]}'
                    else: tempString += f'{foodName} can be stored in the {info} for {min} to {max} {df[info + "_Metric"].iloc[index]}'
                if i == "DOP":
                    min = df['DOP_' + info + '_Min'].iloc[index]
                    max = df['DOP_' + info + '_Max'].iloc[index]
                    if min == max: tempString += f'The Date of Perish for {foodName} is {int(min)} {df["DOP_" + info + "_Metric"].iloc[index]}'
                    else: tempString += f'The Date of Perish for {foodName} is from {int(min)} to {int(max)} {df["DOP_" + info + "_Metric"].iloc[index]}'
                if i == "After":
                    if i == "Thawing":
                        min = df[info + '_After_Thawing_Min'].iloc[index]
                        max = df[info + '_After_Thawing_Max'].iloc[index]
                        if min == max: tempString += f' can be store in the pantry for {min} {df[info + "_After_Thawing_Metric"].iloc[index]}'
                        else: tempString += f' can be store in the pantry for {min} and {max} {df[info + "_After_Thawing_Metric"].iloc[index]}'
                    else:
                        min = df[info + '_After_Opening_Min'].iloc[index]
                        max = df[info + '_After_Opening_Max'].iloc[index]
                        if min == max: tempString += f' can be store in the pantry for {min} {df[info + "_After_Opening_Metric"].iloc[index]}'
                        else: tempString += f' can be store in the pantry for {min} and {max} {df[info + "_After_Opening_Metric"].iloc[index]}'
                if i == "tips" or i == "Tips":
                    if info == "Freeze":
                        for v in col: 
                            if i == "DOP": tempString += f'A tip about {foodName} the Date of Perish is {df["DOP_" + info + "_Tips"].iloc[index]}'
                        tempString += f'A tip about {foodName} is that {df[info + "_Tips"].iloc[index]}'
                    else:
                        for v in col: 
                            if i == "DOP": tempString += f'A {foodName} tip about the Date of Perish is {df["DOP_" + info + "_tips"].iloc[index]}'
                        tempString += f'A tip about {foodName} is that {df[info + "_tips"].iloc[index]}'
                        
    return tempString if tempString != '' else 'No Tip to give!'
```
